```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
urls_arquivos_csv = [
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_Aave.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_BinanceCoin.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_Bitcoin.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_ChainLink.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_Cosmos.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_EOS.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_Ethereum.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_Polkadot.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_USDCoin.csv',
    'https://raw.githubusercontent.com/mathuscm/Criptomoedas-m5/main/Arquivos%20Crypto/coin_XRP.csv'
]
```

```python
dados_moedas = {}

for url in urls_arquivos_csv:
    # Extrai o nome da moeda do URL
    nome_moeda = url.split('/')[-1].replace('coin_', '').replace('.csv', '')

    # Carrega o arquivo CSV em um DataFrame
    df = pd.read_csv(url)

    # Armazena o DataFrame no dicion√°rio usando o nome da moeda como chave
    dados_moedas[nome_moeda] = df
```

```python
# Plot dos valores de todas as criptomoedas
plt.figure(figsize=(12, 6))
for moeda, df in dados_moedas.items():
    plt.plot(df['Date'], df['Close'], label=moeda)

plt.xlabel('Data')
plt.ylabel('Valor de Fechamento')
plt.title('Valores de Fechamento de Criptomoedas ao Longo do Tempo')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Dicion√°rio para armazenar a maior varia√ß√£o percentual de pre√ßo para cada moeda
maior_variacao = {'Moeda': None, 'Data': None, 'Varia√ß√£o Percentual': None}

# Iterar pelas moedas
for moeda, df in dados_moedas.items():
    # Calcular a varia√ß√£o percentual de pre√ßo para cada dia
    df['Varia√ß√£o Percentual'] = (df['Close'] / df['Open'] - 1) * 100
    # Encontrar o dia com a maior varia√ß√£o percentual
    dia_maior_variacao = df.loc[df['Varia√ß√£o Percentual'].idxmax()]

    # Se a varia√ß√£o encontrada for maior do que a armazenada, atualizar o dicion√°rio
    if maior_variacao['Varia√ß√£o Percentual'] is None or dia_maior_variacao['Varia√ß√£o Percentual'] > maior_variacao['Varia√ß√£o Percentual']:
        maior_variacao['Moeda'] = moeda
        maior_variacao['Data'] = dia_maior_variacao['Date']
        maior_variacao['Varia√ß√£o Percentual'] = dia_maior_variacao['Varia√ß√£o Percentual']

# Exibir a moeda com a maior varia√ß√£o percentual em um √∫nico dia
print(f'A criptomoeda com a maior varia√ß√£o percentual de pre√ßo em um √∫nico dia durante o per√≠odo analisado √© {maior_variacao["Moeda"]}.')
print(f'Isso ocorreu em {maior_variacao["Data"]} com uma varia√ß√£o de {maior_variacao["Varia√ß√£o Percentual"]:.2f}%.')

```

```python
# Dicion√°rio para armazenar as maiores varia√ß√µes percentuais
maiores_variacoes = {'Moeda': [], 'Varia√ß√£o Percentual': []}

# Iterar pelas moedas
for moeda, df in dados_moedas.items():
    # Encontrar o dia com a maior varia√ß√£o percentual
    dia_maior_variacao = df.loc[df['Varia√ß√£o Percentual'].idxmax()]
    # Adicionar a moeda e a varia√ß√£o ao dicion√°rio
    maiores_variacoes['Moeda'].append(moeda)
    maiores_variacoes['Varia√ß√£o Percentual'].append(dia_maior_variacao['Varia√ß√£o Percentual'])

# Criar o gr√°fico de barras
plt.figure(figsize=(12, 6))
plt.bar(maiores_variacoes['Moeda'], maiores_variacoes['Varia√ß√£o Percentual'])
plt.xlabel('Moeda')
plt.ylabel('Varia√ß√£o Percentual (%)')
plt.title('Maior Varia√ß√£o Percentual de Pre√ßo em um √önico Dia por Moeda')
plt.xticks(rotation=45)

# Exibir o gr√°fico
plt.tight_layout()
plt.show()
```
