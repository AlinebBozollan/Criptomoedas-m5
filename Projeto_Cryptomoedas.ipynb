```python
# Importanto bibliotecas
import pandas as pd
```

```python
# Extrair arquivo
aave = pd.read_csv("coin_Aave.csv")
aave
```

```python
binance = pd.read_csv("coin_BinanceCoin.csv")
binance
```

```python
bitcoin = pd.read_csv("coin_Bitcoin.csv")
bitcoin
```

```python
chainlink = pd.read_csv("coin_ChainLink.csv")
chainlink
```

```python
cosmos = pd.read_csv("coin_Cosmos.csv")
cosmos
```

```python
eos = pd.read_csv("coin_EOS.csv")
eos
```

```python
ethereum = pd.read_csv("coin_Ethereum.csv")
ethereum
```

```python
polkadot = pd.read_csv("coin_Polkadot.csv")
polkadot
```

```python
usd = pd.read_csv("coin_USDCoin.csv")
usd
```

```python
xrp = pd.read_csv("coin_XRP.csv")
xrp
```

```python
# Concatenando todas as tabelas de criptomoedas
crypto_coins = pd.concat([aave, binance, bitcoin, chainlink, cosmos, eos, ethereum, polkadot, usd, xrp], ignore_index= True)
crypto_coins
```

```python
# convertendo formato da data
crypto_coins['Date'] = pd.to_datetime(crypto_coins['Date']).dt.strftime('%d-%m-%Y')
```

```python
# renomear colunas
crypto_coins.rename(columns={'Name': 'Nome', 'Symbol': 'Símbolo', 'Date':'Data', 'High': 'Alto', 'Low': 'Baixo', 'Open': 'Aberto', 'Close': 'Fechado', 'Marketcap': 'Valor da ação'}, inplace=True)
```

```python
crypto_coins.drop(["SNo"], axis = 1, inplace = True)
```

```python
crypto_coins
```

```python
# convertendo dataframe para csv
crypto_coins.to_csv("crypto_coins.csv")
```
